```python
import requests
import json

req = requests.get("https://github.com/bhancockio/langchain-crash-course/blob/main/4_rag/books/adventures_of_huckleberry_finn.txt")

# Check if the request was successful
if req.status_code == 200:
  # Access the content as text
  resp = req.content.decode('utf-8')
  print(resp)
else:
  print(f"Request failed with status code: {req.status_code}")
```
